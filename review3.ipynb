```python
x = [250, 10, 3, 150,250]
y = [100, 25, 150, 0,100]

x1 = [50, 210, 13, 10,20]
y1 = [100, 0, 150, 20,100]
plt.plot(x,y, "mx-", label='Data1')
plt.plot(x1,y1, "co:",label='Data-1')
plt.xlabel('Truc x')
plt.ylabel('Truc y')
plt.show()
```

```python
type = ['A', 'B', 'C', 'D','E']
values = [250, 150, 30, 150,250]

plt.bar(type, values,color="red", alpha =0.6)
plt.show()
```

```python
values = [20, 10, 30, 15,25]
labels  = ['A', 'B', 'C', 'D','E']
colors = ['red', 'green', 'blue', 'gray','violet']

plt.pie(values, labels=labels , colors=colors,autopct='%1.1f%%')
plt.title('Test title')
plt.show()
```

```python
values = [20, 35, 30, 50,100, 76, 98, 23]

plt.boxplot(values)
plt.title("Bieu do hop")
plt.ylabel=('gia tri cua hop')
plt.show()
```

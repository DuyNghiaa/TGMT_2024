```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
mydir = 'C:/Users/user/Documents/Th gic my tnh/image/1.jpg'
im = Image.open(mydir)
plt.imshow(im)
```

```python
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title("Bieu do histogram")
plt.xlabel("value")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```

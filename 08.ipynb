```python
import matplotlib.pyplot as plt
```

```python
x =[1,3,5,7,9]
y= [10, 20, 10, 40, 30]

x1=[2,4,6,8,19]
y1=[50,40,30,20,10]

plt.plot(x, y ,'go-', label="Data") #draw 1
plt.plot(x1, y1, marker='o', color='r', linestyle ='-', label='Data 1')#draw 2
plt.legend()
plt.xlabel('Truc x')
plt.ylabel('Truc y')
plt.show()
```

```python
categories =['A', 'B', 'C', 'D', 'E']
values =  [15, 10, 15, 18, 20]

plt.bar(categories, values, color='g', alpha=0.6) # V biu  ct
plt.xlabel('Danh mc') # t nhn cho trc x
plt.ylabel('Gi tr') # t nhn cho trc y
plt.show() # Hin th biu 
```

```python
my_labels = ['A', 'Y', 'E', 'L']
sizes = [15, 20, 45, 20]
my_colors = ['red', 'green', 'blue', 'orange']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%') # V biu  hnh trn
plt.title('Biu  hnh trn')
plt.show()
```

```python
x =[1,3,5,7,9]
y= [10, 20, 10, 40, 30]
plt.scatter(x, y, c='r', marker='o', label='Data')  # V biu  phn tn
plt.title('Biu  phn tn v d')  # t tiu  cho biu 
plt.xlabel('Trc x')  # t nhn cho trc x
plt.ylabel('Trc y')  # t nhn cho trc y
plt.legend()  # Hin th ch gii
plt.show()  # Hin th biu 
```

```python
data = [15, 18, 12, 20, 35, 15, 50, 55, 65]

plt.boxplot(data)  # V biu  hp
plt.title('Biu  hp v d')  # t tiu  cho biu 
plt.ylabel('Gi tr')  # t nhn cho trc y
plt.show()  # Hin th biu 
```

```python
data = [15, 18, 22, 10, 35, 25, 50, 55, 25]

plt.violinplot(data)  # V biu  violin
plt.title('Biu  violin v d')  # t tiu  cho biu 
plt.ylabel('Gi tr')  # t nhn cho trc y
plt.show()  # Hin th biu 
```

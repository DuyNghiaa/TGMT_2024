```python
import matplotlib.pyplot as plt
```

```python
x =[1,3,5,7,9]
y= [10, 20, 10, 40, 30]

x1=[2,4,6,8,19]
y1=[50,40,30,20,10]

plt.plot(x, y ,'go-', label="Data") #draw 1
plt.plot(x1, y1, marker='o', color='r', linestyle ='-', label='Data 1')#draw 2
plt.legend()
plt.xlabel('Truc x')
plt.ylabel('Truc y')
plt.show()
```

```python
categories =['A', 'B', 'C', 'D', 'E']
values =  [15, 10, 15, 18, 20]

plt.bar(categories, values, color='g', alpha=0.6) # Vẽ biểu đồ cột
plt.xlabel('Danh mục') # Đặt nhãn cho trục x
plt.ylabel('Giá trị') # Đặt nhãn cho trục y
plt.show() # Hiển thị biểu đồ
```

```python
my_labels = ['A', 'Y', 'E', 'L']
sizes = [15, 20, 45, 20]
my_colors = ['red', 'green', 'blue', 'orange']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%') # Vẽ biểu đồ hình tròn
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python
x =[1,3,5,7,9]
y= [10, 20, 10, 40, 30]
plt.scatter(x, y, c='r', marker='o', label='Data')  # Vẽ biểu đồ phân tán
plt.title('Biểu đồ phân tán ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.xlabel('Trục x')  # Đặt nhãn cho trục x
plt.ylabel('Trục y')  # Đặt nhãn cho trục y
plt.legend()  # Hiển thị chú giải
plt.show()  # Hiển thị biểu đồ
```

```python
data = [15, 18, 12, 20, 35, 15, 50, 55, 65]

plt.boxplot(data)  # Vẽ biểu đồ hộp
plt.title('Biểu đồ hộp ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.ylabel('Giá trị')  # Đặt nhãn cho trục y
plt.show()  # Hiển thị biểu đồ
```

```python
data = [15, 18, 22, 10, 35, 25, 50, 55, 25]

plt.violinplot(data)  # Vẽ biểu đồ violin
plt.title('Biểu đồ violin ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.ylabel('Giá trị')  # Đặt nhãn cho trục y
plt.show()  # Hiển thị biểu đồ
```
